```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
import numpy as np
import re
```

##### Search Condition : 更新日期一個月內、有寫薪資、經驗要求一年以下、學歷要求大學碩士


```python
Job_Info = [] ; Company_Info = [] ; Tag = [] ; Description = []
for i in range(150):
    print(i)
    url = 'https://www.104.com.tw/jobs/search/?ro=1&order=11&asc=0&sr=99&excludeCompanyByCustno=7e2s7e9%2C1a2x6bl0ns%2C1a2x6birjm&excludeIndustryCat=1016000000%2C1009001000%2C1009005000%2C1009004000%2C1005001001&page='+ str(i+1) +'&jobexp=1&mode=s&jobsource=2018indexpoc'
    rep = requests.get(url)
    doc = BeautifulSoup(rep.text, 'html.parser')
    temp = doc.find_all(['article'])
    
    for nodes in temp:
        ID = str(np.random.randint(0,9))+ str(np.random.randint(1000000,9999999)) 

        
        # Create Job_info list
        job_info = nodes.find_all('a')
        job = [ID]
        for j in job_info:
            job.append(j.text.replace("\n","").replace(' ',''))
        Job_Info.append(job)
        
        
        # Create Company_Info list
        c_info = nodes.find_all('li')
        company_info = [ID]
        for j in c_info:
            company_info.append(j.text.replace(' ','').replace('\n',''))
        Company_Info.append(company_info)
        
        
        # Create Tag list
        tags  = nodes.find_all('span', class_ = 'b-tag--default')
        tag = [ID]
        for j in tags:
            tag.append(j.text)
        Tag.append(tag)
        
        
        # Create Description list'
        texts = nodes.find_all('p',class_ = "job-list-item__info b-clearfix b-content")
        description = [ID]
        for j in texts:
            description.append(j.text.replace('\n', '').replace('\r','').replace('\u3000',''))
        Description.append(description)

        
print(f"Shape of Job_Info : {len(Job_Info)}")        
print(f"Shape of Company_Info : {len(Company_Info)}")
print(f"Shape of Tag : {len(Tag)}")
print(f"Shape of Description : {len(Description)}")

        
```

```python
job_info = Job_Info
company_info = Company_Info
tag = Tag
description = Description
```

```python
# Check element sizes
for i in range(len(description)):
    print(f"Job_Info : {len(job_info[i])}", f"Company_Info : {len(company_info[i])}", f"Tag : {len(tag[i])}", f"Description : {len(description[i])}")
    
```

```python
# Align tag list 
for i in tag:
    if len(i) != 4:
        n = 4-len(i)
        for j in range(n):
            i.append(np.NaN)
```

```python
# Drop Failed elements
i = 0
while i < len(job_info):
    if len(job_info[i]) !=4:
        del job_info[i]
        del company_info[i]
        del tag[i]
        del description[i]
    else:
        i+=1
```

```python
# rearrange company_info data
for k in range(len(company_info)):
    temp = company_info[k]
    
    # ID, name, industry, loc, exp, edu
    if len(temp) == 6:
        temp[0:6] = temp[0], temp[2], np.NaN, temp[3], temp[4], temp[5]
    else:
        temp[0:6] = temp[0], temp[2], temp[3], temp[4], temp[5], temp[6]
        temp = temp[0:6]
    company_info[k] = temp
    
```

```python
# rearrange tag data
for i in tag:
    if (type(i[-1]) == float) & (type(i[-2]) == float):   
        pass
    
    elif (type(i[-1])== float) & ("人" in i[-2]):
        i[2], i[-1] = i[-1], i[2]
        
    else:
        pass
```

```python
df_job_info = pd.DataFrame(job_info, columns = ['ID','Title','Company','app_num'])
df_company_info = pd.DataFrame(company_info, columns = ["ID","Company","Industry","Location","Experience",'Edu'])
df_tag = pd.DataFrame(tag, columns=["ID","Pay","Tag","Size"])
df_descript = pd.DataFrame(Description, columns=["ID","text"])
```

```python
df = df_company_info.merge(df_job_info)
df = df.merge(df_tag)
df = df.merge(df_descript)
df.head()
```

```python
drop = []
for i in range(len(df.Pay)):
    if '論件計酬' in df.Pay[i]:
        drop.append(i)
df.drop(index=drop, inplace = True)
```

```python
# Get the total employee number 
pattern = "([0-9-])+"
num = [int(re.search(pattern, x).group(0)) if pd.notna(x) else np.NaN for x in df.Size]
df['Employee_Num'] = num


# Get pay type
pattern = ".(薪|元)"
df["Pay_type"] = df.Pay.map(lambda x: re.search(pattern, x).group(0))


# Get pay num
pattern = "薪(.)*元"
salary = df.Pay.map(lambda x: re.search(pattern, x).group(0).replace('薪','').replace('元',''))
df[['Salary_floor', "Salary_ceiling"]] = salary.str.replace(',','').str.split('~', expand = True).fillna(0).astype(int)

# Normalize to monthly paid
df["Salary_floor"] = df.apply(lambda x: x.Salary_floor/12 if x.Pay_type == '年薪' else x.Salary_floor, axis=1)
df["Salary_ceiling"] = df.apply(lambda x: x.Salary_ceiling/12 if x.Pay_type == '年薪' else x.Salary_ceiling, axis=1)


# transform app_num
pattern = "[0-9-~]*"
num = df.app_num.map(lambda x:re.search(pattern,x).group(0))
df['app_status'] = num.map({'30':'Hot','11~30':'Above_Avg','6~10':'Below_Avg','0~5':'Cold'})
```

```python
df.to_csv('104_fresh_employee_all.csv', index=False)
```
